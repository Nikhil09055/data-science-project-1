```python
import pandas as pd
data1 = {'Name' : ['Nikki','Ishu','Hari','Mani'],
         'Age' : [27,24,22,32],
         'Address': ['Maddipadu','Maddipadu','Kanigiri','Ongole'],
         'Qualification': ['MSC','MSC','MCA','PHD']}

data2 = {'Name' : ['Naga','Yashu','Moksha','Sai'],
         'Age' : [23,24,27,32],
         'Address': ['Hyderbad','Hyderbad','Banglore','Mysore'],
         'Qualification': ['B-TECH','Bcom','Phd','B.hons']}

df = pd.DataFrame(data1,index=[0,1,2,3])
df1 = pd.DataFrame(data2,index=[4,5,6,7])
print(df,"\n\n",df1)
res1 = pd.concat([df,df1])
res1
print(df,"\n\n",df1)
res2=pd.concat([df,df1],axis=1,join='inner')
res2
```
