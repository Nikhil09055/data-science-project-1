```python
from matplotlib import pyplot as plt
import numpy as np
import math
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel("Angle")
plt.ylabel("Sine")
plt.title("Sine wave")
plt.show()
```

```python
import matplotlib.pyplot as plt
year=[1,2,3]
Veg_Rate=[2,4,6]
Milk_Rate=[1,3,5]
Fruit_Rate=[3,6,9]
line1,=plt.plot(year,Veg_Rate)
line2,=plt.plot(year,Milk_Rate)
line3,=plt.plot(year,Fruit_Rate)
plt.legend([line1, line2, line3],['Vegetable','Milk','Fruit'])
plt.xlabel('years');
plt.ylabel('Rate');
plt.show()
print("successfully placed a legend on the axis..")
```

```python
import matplotlib.pyplot as plt
categories = ['Lion', 'Horse', 'Dog']
values = [15,24,30]
plt.bar(categories, values, color='black')
plt.xlabel('categories')
plt.ylabel('values')
plt.title("Basic vertical bar graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
categories = ['Lion', 'Horse', 'Dog']
values1=[15,24,30]
values2=[20,18,25]

bar_width=0.35
bar_position1=np.arange(len(categories))
bar_position2=bar_position1+bar_width
plt.bar(bar_position1, values1, width=bar_width, label='2021', color='red')
plt.bar(bar_position2, values2, width=bar_width, label='2022', color='yellow')
plt.xlabel('categories')
plt.ylabel('values')
plt.title("Grouped bar garph")
plt.legend()
plt.show()
```
